```python
def accuracy(y_test, y_pred):
    # Initialize the count of correct guesses
    correct_guess = 0

    # Ensure y_test and y_pred have the same size
    if len(y_test) != len(y_pred):
        raise ValueError("y_test and y_pred must have the same size")

    # Iterate through the elements of y_test and y_pred
    for actual, predicted in zip(y_test, y_pred):
        if predicted - 1 <= actual <= predicted + 1:
            correct_guess += 1

    # Calculate and return the accuracy as a floating-point number
    accuracy = correct_guess / len(y_test)
    return accuracy
```
