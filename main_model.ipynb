```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cards = pd.read_csv("Credit Cards.csv").dropna()
cards
```

```python
columns_to_drop = ['Name','Min_age','Min_income','Credit_score','Max_age','Features']
card_clean=cards.drop(columns=columns_to_drop)
card_clean['Freq']=1
card_clean
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

columns_to_scale = ["Interest", "Fees"]

card_clean[columns_to_scale] = scaler.fit_transform(card_clean[columns_to_scale])
card_clean.head()
```

```python
intercept_ss = 3
default_wt = np.array([-0.75, 2.5 ,-0.1 ,-0.2 ,-0.2 ,-0.1 ,0.75 ,0.2,0.25 ,1.0,0.2 ,0.5 ,0.2 ,0.2 ,0.5 ,0.3, 0.25 ])
default_wt.shape
```

wt[6] = fuel,
wt[9] = shopping,
wt[11] = travel,
wt[14] = cashback,
wt[15] = entertainment,
wt[16] = freq

```python

def top_cards_by_default(n):
  data = cards.copy()
  wt = default_wt.copy()
  wt[16]=2*wt[16]
  card_array=card_clean.drop(columns=['ID'])
  data['Score'] = 2 + card_array.dot(wt)
  order=data[['Name','Score']]
  order = order.sort_values(by='Score', ascending=False)
  print(f"Our {n} top credit cards for average user: ")
  print(order.head(n))
```

```python
n = input("Enter a number of cards to be displayed: ")
n = int(n)
top_cards_by_default(n)
```

```python
def satisfaction_score(isFuel,isShp,isTr,isCb,isEn,Fr,ID, isInt):
  wt = default_wt.copy()
  if isInt:
    wt[0]=0.2*wt[0]
  if isFuel:
    wt[6]=2*wt[6]
  if isShp:
    wt[9]=2*wt[9]
  if isTr:
    wt[11]=2*wt[11]
  if isCb:
    wt[14]=2*wt[14]
  if isEn:
    wt[15]=2*wt[15]
  wt[16]=Fr*wt[16]

  arr = card_clean[card_clean['ID'] == ID]
  val = arr.to_numpy()
  val=val.T
  val=val[1:]
  wt_reshaped=wt.reshape(-1,1).T
  dp  = intercept_ss + wt_reshaped.dot(val)
  lower = max(1,round(dp[0][0],1)-0.5)
  upper = min(10,round(dp[0][0],1)+0.5)
  print("\nPredicted score is ",dp[0][0])
  print(f"\nYour expected Satisfaction score is between {lower} and {upper}")
```

```python
def get_satisfaction_score():
  for index, row in cards.iterrows():
    print(f"{row['ID']}: {row['Name']}",end="  ")
  id=int(input("\nSelect a card based on the card ID: "))
  name = cards[cards['ID']==id]['Name'].values[0]
  print("You are using ",name)
  isInt = int(input("Do you pay your bills before the due date? '1' for Yes, '0' for No: "))
  print("\nHow frequently do you use your credit card?")
  print("1: Very Rare")
  print("2: Sometimes")
  print("3: Very frequently")
  freq=int(input())
  print("\nFor what purposes do you use your credit card: '1' for Yes, '0' for No")
  isShop = int(input("Online Shopping? "))
  isTr = int(input("Travel and Tickets? "))
  isFuel = int(input("Fuel? "))
  isCb = int(input("Cashback offers? "))
  isEn = int(input("Entertainment(Movies,etc)? "))
  satisfaction_score(isFuel,isShop,isTr,isCb,isEn,freq,id,isInt)
```

```python
get_satisfaction_score()
```

```python
def recommender():
  age = int(input("Enter your age: "))
  if age < 18 :
    print("Sorry no available cards for you :)")
    return
  income = int(input("Enter your income in LPA: "))
  credit_score=int(input("Enter your credit score (Some companies want credit score of their users to be higher. Write 0 if you dont want to reveal): "))
  eligible_cards = cards[
    (cards['Min_age'] <= age) &
    (cards['Max_age'] >= age) &
    (cards['Min_income'] <= income) &
    (cards['Credit_score'] <= credit_score)
  ]
  eligible_ids = eligible_cards['ID'].tolist()
  if(len(eligible_cards)==0):
    print("Sorry no available cards for you :)")
    return
  isInt = int(input("Do you pay your bills before the due date? '1' for Yes, '0' for No: "))
  print("\nHow frequently will you use your credit cards?")
  print("1: Very Rare")
  print("2: Sometimes")
  print("3: Very frequently")
  Fr=int(input())
  print("\nFor what purposes will you use your credit card: '1' for Yes, '0' for No")
  isShp = int(input("Online Shopping? "))
  isTr = int(input("Travel and Tickets? "))
  isFuel = int(input("Fuel? "))
  isCb = int(input("Cashback offers? "))
  isEn = int(input("Entertainment(Movies,etc)? "))
  wt = default_wt.copy()
  if isInt:
    wt[0]=0.2*wt[0]
  wt[16]=Fr*wt[16]
  if isFuel:
    wt[6]=2*wt[6]
  if isShp:
    wt[9]=2*wt[9]
  if isTr:
    wt[11]=2*wt[11]
  if isCb:
    wt[14]=2*wt[14]
  if isEn:
    wt[15]=2*wt[15]

  card_array=card_clean.drop(columns=['ID'])
  score = card_array.dot(wt)
  result = cards[['ID', 'Name']].copy()
  result['score'] = score
  result = result.sort_values(by='score', ascending=False)
  eligible_cards = result[result['ID'].isin(eligible_ids)]
  final_cards = eligible_cards['Name'].tolist()
  m = len(final_cards)
  if(m==0):
    print("Sorry no available cards for you :)")
  else:
    n = int(input("How many cards do you want us to recommend: "))
    print("\nRecommended cards is/are :")
    for i in range(min(m,n)):
      print(final_cards[i])
```

```python
recommender()
```
